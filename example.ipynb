```python
from color_transfer import color_transfer
import numpy as np
import argparse
import cv2
import os
from IPython.display import Image
```

Path to the source image

```python
source_path = os.path.join('images', 'autumn.jpg')
```

Path to the target image

```python
target_path = os.path.join('images', 'fallingwater.jpg')
```

Should np.clip scale L*a*b* values before final conversion to BGR?<br>
Approptiate min-max scaling used if False.

```python
clip = True
```

Should color transfer strictly follow methodology layed out in original paper?

```python
preserve_paper = True
```

Path to the output image (optional)

```python
transfered_path = os.path.join('images', 'transfered.jpg')
```

load the images

```python
source = cv2.imread(source_path)
target = cv2.imread(target_path)
```

transfer the color distribution from the source image<br>
to the target image

```python
transfer = color_transfer(source, target, clip, preserve_paper)
```

check to see if the output image should be saved

```python
cv2.imwrite(transfered_path, transfer)
```
